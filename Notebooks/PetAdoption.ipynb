```python
from sklearn import model_selection

def create_folds(data,label):
    data['kfold'] = -1
    data = data.sample(frac = 1).reset_index(drop=True)
    num_bins = int(np.floor(1+np.log2(len(data))))
    
    data.loc[:,"bins"] = pd.cut(data[label],bins = num_bins, labels = False)
    
    
    kf = model_selection.StratifiedKFold(n_splits = 10)
    
    for f,(t_,v_) in enumerate(kf.split(X=data,y=data.bins.values)):
        data.loc[v_,'kfold'] = f
    data = data.drop("bins", axis = 1)
    return data
```

```python
df1 = create_folds(train, 'breed_category')
df = create_folds(train, 'pet_category')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score
from xgboost import XGBClassifier
```

```python
print("DecisionTree")
train_acc = []
test_acc = []
for i in range(1):
    trainX = df1[df1["kfold"]!=i].drop(['pet_id','issue_date','listing_date','pet_category',"breed_category",'kfold'],axis = 1)
    trainY = df1[df1["kfold"]!=i][['breed_category']]
    
    testX = df1[df1["kfold"]==i].drop(['pet_id','issue_date','listing_date','pet_category',"breed_category",'kfold'],axis = 1)
    testY = df1[df1["kfold"]==i][['breed_category']]
    
    
    depth = 12
    
    clfTree = XGBClassifier()
    clfTree.fit(trainX,trainY)
    
    train_acc.append(f1_score(clfTree.predict(trainX),trainY, average='weighted'))
    test_acc.append(f1_score(clfTree.predict(testX),testY, average='weighted'))
    
    
    prediction = clfTree.predict(testX)
#     print(f"Train Accuracy for fold No. {i} for depth {depth}: ",f1_score(clfTree.predict(trainX),trainY, average='weighted'))
#     print(f"Test Accuracy for fold No. {i}  for depth {depth}: ",f1_score(clfTree.predict(testX),testY, average='weighted'))
print('train avg',sum(train_acc)/len(train_acc))
print('test avg',sum(test_acc)/len(test_acc))
```

```python
predY = clfTree.predict(test.drop(['pet_id','issue_date','listing_date',"pet_category","breed_category"],axis = 1))
test['breed_category'] = predY
test
```
