```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("../Dataset/train.csv")
test = pd.read_csv("../Dataset/test.csv")
```

```python
test["breed_category"]=-1
test["pet_category"]=-1
test.head()
```

```python
print(f"condition column has following types :\n {train['condition'].value_counts()}\n")
print(f"color_type column has following types :\n {train['color_type'].value_counts()}\n")
print(f"X1 column has following types :\n {train['X1'].value_counts()}\n")
print(f"X2 column has following types :\n {train['X2'].value_counts()}\n")
print(f"breed_category column has following types :\n {train['breed_category'].value_counts()}\n")
print(f"pet_category column has following types :\n {train['pet_category'].value_counts()}\n")
```

```python
df = pd.concat([train,test]).reset_index(drop = True)
```

```python
df['listing_date'] = pd.to_datetime(df['listing_date'])
df['issue_date'] = pd.to_datetime(df['issue_date'])
df['days_taken'] = df['listing_date'] - df['issue_date']
df['days_taken'] = df['days_taken'].dt.days
df['breed_category'] = df['breed_category'].astype(int)
```

```python
from sklearn import preprocessing

features = [x for x in train.columns if x not in [ 'pet_id','issue_date','listing_date','length(m)', 'height(cm)'
       'days_taken','pet_category','breed_category']]

for feat in features:
    lbl_enc = preprocessing.LabelEncoder()
   
    temp_col = df[feat].fillna("None").astype(str).values
    
    df.loc[:,feat] = lbl_enc.fit_transform(temp_col)
    
    
df['X1*X2'] = df['X1']*df['X2']
df['X1+X2'] = df['X1']+df['X2']
df['X1-X2'] = df['X1']-df['X2']

df['size'] = df['length(m)']*df['height(cm)']/100

df['issue_month'] = df['issue_date'].dt.month

df['listing_month']=df['listing_date'].dt.month

df['issue_year'] = df['issue_date'].dt.year

df['listing_year']=df['listing_date'].dt.year

df['length(m)'] = (df['length(m)']-df['length(m)'].min())/(df['length(m)'].max()-df['length(m)'].min())

df['height(cm)'] = (df['height(cm)']-df['height(cm)'].min())/(df['height(cm)'].max()-df['height(cm)'].min())

df['size'] = (df['size']-df['size'].min())/(df['size'].max()-df['size'].min())

df['days_taken'] = (df['days_taken']-df['days_taken'].min())/(df['days_taken'].max()-df['days_taken'].min())
df = pd.get_dummies(df,columns=['condition','X1','X2','color_type','listing_month','issue_month','listing_year','issue_year'],prefix=['condition','X1','X2','color_type','listing_month','issue_month','listing_year','issue_year'])
# df = pd.get_dummies(df,columns=['condition','X1','X2'],prefix =['condition','X1','X2'] )  
train = df[df.breed_category != -1].reset_index(drop = True)
test = df[df.breed_category == -1].reset_index(drop = True)
```

```python
from sklearn import model_selection

def create_folds(data,label):
    data['kfold'] = -1
    data = data.sample(frac = 1).reset_index(drop=True)
    num_bins = int(np.floor(1+np.log2(len(data))))
    
    data.loc[:,"bins"] = pd.cut(data[label],bins = num_bins, labels = False)
    
    
    kf = model_selection.StratifiedKFold(n_splits = 10)
    
    for f,(t_,v_) in enumerate(kf.split(X=data,y=data.bins.values)):
        data.loc[v_,'kfold'] = f
    data = data.drop("bins", axis = 1)
    return data
```

```python
df1 = create_folds(train, 'breed_category')
df = create_folds(train, 'pet_category')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score
from xgboost import XGBClassifier
```

```python
i = 0
trainX = df[df["kfold"]!=i].drop(['pet_id','issue_date','listing_date',"pet_category","breed_category","kfold"],axis = 1)
trainY = df[df["kfold"]!=i][['pet_category']]
    
testX = df[df["kfold"]==i].drop(['pet_id','issue_date','listing_date',"pet_category","breed_category","kfold"],axis = 1)
testY = df[df["kfold"]==i][['pet_category']]
    
depth = 15
clfTree =XGBClassifier(random_state=42)
clfTree.fit(trainX,trainY.values)
```

```python
print("DecisionTree")
for i in range(2):
    trainX = df[df["kfold"]!=i].drop(['pet_id','issue_date','listing_date',"pet_category","breed_category","kfold"],axis = 1)
    trainY = df[df["kfold"]!=i][['pet_category']]
    
    testX = df[df["kfold"]==i].drop(['pet_id','issue_date','listing_date',"pet_category","breed_category","kfold"],axis = 1)
    testY = df[df["kfold"]==i][['pet_category']]
    
    depth = 15
    clfTree = XGBClassifier()
    clfTree.fit(trainX,trainY.values)
    
    prediction = clfTree.predict(testX)
    print(f"Train Accuracy for fold No. {i} for depth {depth}: ",f1_score(clfTree.predict(trainX),trainY, average='weighted'))
    print(f"Test Accuracy for fold No. {i}  for depth {depth}: ",f1_score(clfTree.predict(testX),testY, average='weighted'))
```

```python
predY = clfTree.predict(test.drop(['pet_id','issue_date','listing_date',"pet_category","breed_category"],axis = 1))
test['pet_category'] = predY
test
```

```python
print("DecisionTree")
for i in range(1):
    trainX = df[df["kfold"]!=i].drop(['pet_id','issue_date','listing_date','pet_category',"breed_category",'kfold'],axis = 1)
    trainY = df[df["kfold"]!=i][['breed_category']]
    
    testX = df[df["kfold"]==i].drop(['pet_id','issue_date','listing_date','pet_category',"breed_category",'kfold'],axis = 1)
    testY = df[df["kfold"]==i][['breed_category']]
    
    
    depth = 11
    clfTree = XGBClassifier()
    clfTree.fit(trainX,trainY)
    
    prediction = clfTree.predict(testX)
    print(f"Train Accuracy for fold No. {i} for depth {depth}: ",f1_score(clfTree.predict(trainX),trainY, average='weighted'))
    print(f"Test Accuracy for fold No. {i}  for depth {depth}: ",f1_score(clfTree.predict(testX),testY, average='weighted'))
```

```python
predY = clfTree.predict(test.drop(['pet_id','issue_date','listing_date',"pet_category","breed_category"],axis = 1))
test['breed_category'] = predY
test
```
